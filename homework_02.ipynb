## Задание 1: Получение однозначного числа

Пользователь вводит целое число, программа складывает все цифры числа, с полученным числом — то же самое, и так до тех пор, пока не получится однозначное число.

<br>

**Примеры:**
* `545   -> 5`
* `12345 -> 6`

````python
# ВАШ КОД ЗДЕСЬ
str_num = ""
while not str_num.isdigit():
   str_num = input("Введите число: ")
   num_sum = [int(i) for i in str_num]

while len(num_sum) > 1:
   num_sum = [int(i) for i in str(sum(num_sum))]
else:
   num_sum = str(num_sum.pop(0))

print(str_num, num_sum, sep = " -> ")
````

## Задание 2: Кинотеатр

Дан список списков, каждый вложенный список состоит из 1 и 0, количество вложенных списков — количество рядов. Пользователь вводит, сколько билетов ему требуется. Программа должна найти ряд, где можно приобрести нужное количество билетов (места должны быть рядом). Если таких рядов несколько, то ближайший к экрану (ближайшим считается нулевой ряд). Если таких мест нет, то вывести False.

<br>

**Примеры:**
* `[[0,1,1,0], [1,0,0,0], [0,1,0,0]], 2 -> 1`
* `[[0,1,1,0], [1,0,1,0], [1,1,0,1]], 2 -> False`

````python
# ВАШ КОД ЗДЕСЬ
lst_row = [[0,1,1,0], [1,0,0,0], [0,1,0,0]]
ticket_count = ""

while not ticket_count.isdigit():
   ticket_count = "0" * int(input("сколько билетов требуется?"))

for row_num, row_set in enumerate(lst_row):
   if ticket_count in ''.join(str(x) for x in row_set):
         print(row_num)
         break
else:
   print(False)
````

## Задание 3: Алгоритм RLE

Необходимо написать упрощенную версию алгоритма RLE. 
Алгоритм RLE объединяет подряд идущие символы в коэффициент и символ.

<br>

**Примеры:**
* `asssdddsssddd -> 1a3s3d3s3d`
* `aaabbbbccccc  -> 3a4b5c`
* `abcba         -> 1a1b1c1b1a`

````python
# ВАШ КОД ЗДЕСЬ
str_row = input("Введите строку: ")
char_i = str_row[0]
char_c = 0
str_rle = []

for char in str_row:
   if char_i == char:
      char_c += 1
   else:
      str_rle.append(str(char_c)+char_i)
      char_c = 1
      char_i = char

str_rle.append(str(char_c)+char_i)

str_rle = "".join(x for x in str_rle)

print(str_row, str_rle, sep = " -> ")
````

## Задание 4: Шифр Цезаря

Реализуйте программу для шифрования текста с помощью шифра Цезаря.  
Шифр Цезаря — это метод шифрования, при котором каждая буква в тексте заменяется буквой, стоящей на фиксированное число позиций дальше в алфавите.

### Требования
1. Программа должна принимать на вход:
   - Строку текста для шифрования
   - Ключ шифра (число от 0 до 25)

2. Правила шифрования:
   - Шифруются только латинские буквы (a-z, A-Z)
   - Сохраняется регистр букв (заглавные остаются заглавными)
   - Пробелы, цифры и знаки препинания остаются без изменений
   - Сдвиг циклический (после 'z' идет 'a', после 'Z' идет 'A')

### Примеры

* `"Hello World", 3      -> "Khoor Zruog"`
* `"Python is great!", 5 -> "Udymts nx lwjfy!"`
* `"XYZ abc", 3          -> "ABC def"`
* `"Test 123", 1         -> "Uftu 123"`

### Алгоритм
* Для каждого символа в строке:
   - Если это латинская буква, найти её позицию в алфавите
   - Добавить к позиции значение ключа
   - Если результат больше 25, использовать остаток от деления на 26
   - Преобразовать обратно в букву, сохранив регистр
   - Если это не буква, оставить символ без изменений


````python
# ВАШ КОД ЗДЕСЬ
str_txt = input("Введите строку для шифрования: ")

keycode = ""

while not (0 <= int(keycode) <= 25) if keycode.isdigit() else True:
    keycode = input("Ключ шифра (число от 0 до 25): ")

keycode = int(keycode)

lst_ascii = [[65,91], [97,123]] # минимальные и максимальные интервальные значения ASCII кодов заглавных и строчных букв

encryption = {}

for i in range(len(lst_ascii)):
    encryption.update(dict((chr(ascii_code), chr((ascii_code + keycode) % lst_ascii[i][1] + lst_ascii[i][0]) if ascii_code + keycode >= lst_ascii[i][1] else chr(ascii_code + keycode)) for ascii_code in range(lst_ascii[i][0], lst_ascii[i][1])))

lst_txt = list(str_txt)

for i, char in enumerate(str_txt):
    lst_txt[i] = encryption.get(char, char)

str_encrypted = ''.join(str(x) for x in lst_txt)
print(str_txt, str_encrypted, sep = " -> ")
````

## Задание 5: Табель успеваемости

Пользователь в бесконечном цикле (пока не будет введена пустая строка) вводит строки вида: 'название предмета' 'фамилия ученика' 'оценка'. После окончания ввода программа выводит в консоль название предмета, далее список учеников и все их оценки в виде таблицы.

<br>

**Примеры:**

<br>

*Ввод:*
```python
Математика Иванов  5
Математика Иванов  4
Литература Иванов  3
Математика Петров  5
Литература Сидоров 3
Литература Петров  5
Литература Иванов  4
Математика Сидоров 3
Математика Петров  5
```

<br>

*Результат:*
```python
Математика # вводим название предмета

# выводим список учеников и их оценки
Иванов  5 4
Петров  5 5
Сидоров 3
```

```python
Литература # вводим название предмета

# выводим список учеников и их оценки
Иванов  3 4
Сидоров 3
Петров  5
```

````python
# ВАШ КОД ЗДЕСЬ
lst_predmet = []
lst_familija = []
lst_ocenka = []
lst_baza = []
char_count = 0

while True:
    str_predmet = input("название предмета: ")

    if str_predmet == "":
        break

    str_familija = input("фамилия ученика: ")

    if str_familija == "":
        break

    str_ocenka = input("оценка: ")

    if str_ocenka == "":
        break
    else:
        str_ocenka = int(str_ocenka)

    if str_predmet not in lst_predmet:
        lst_predmet.append(str_predmet)

    if str_familija not in lst_familija:
        lst_familija.append(str_familija)
        char_count = len(str_familija) if len(str_familija) > char_count else char_count

    if str_ocenka not in lst_ocenka:
        lst_ocenka.append(str_ocenka)

    lst_baza.append([lst_predmet.index(str_predmet), lst_familija.index(str_familija), lst_ocenka.index(str_ocenka)])

if len(lst_baza) != 0:
    while True:
        str_predmet = input("Введите запрос в формате 'название предмета': ")
        print("\n" + str_predmet, end="\n\n")

        if str_predmet == "":
            break
        else:
            if str_predmet in lst_predmet:
                for familija in lst_familija:
                    str_rezultat = ""

                    for row in lst_baza:
                        if row[0] == lst_predmet.index(str_predmet) and lst_familija[row[1]] == familija:
                            str_rezultat += str(lst_ocenka[row[2]]) + " "

                    print(familija,str_rezultat, sep=" " * (char_count - len(familija) + 1))
````
