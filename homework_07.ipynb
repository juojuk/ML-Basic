### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
# одномерный массив размера 10
z = np.zeros(10)
print(f"{z}\n")

# пятый элемент равен 1
np.put(z, 4, 1)
print(f"{z}\n")

# Трансформируем в двумерный массив
z = z.reshape(2, -1)
print(f"{z}\n")
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
# одномерный массив со значениями от 10 до 49
a = np.arange(10, 50)
print(f"{a}\n")

# развернем его (первый элемент становится последним)
a = a[::-1]
print(f"{a}\n")

# Находим в нем все четные элементы.
a = a[a % 2 == 0]
print(f"{a}\n")
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
# 
a = np.arange(0, 9).reshape(3, 3)
print(a)
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
#  массив 4x3x2 со случайными значениями
a = np.random.rand(24).reshape(4, 3, 2)
print(f"{a}\n")

# Находим его минимум и максимум
print("Минимум:", a.min())
print("Максимум:", a.max())
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
# два двумерных массива размерами 6x4 и 4x3
a = np.random.rand(24).reshape(6, 4)
b = np.random.rand(12).reshape(4, 3)

print(f"{a}\n")
print(f"{b}\n")

# Находим их матричное произведение #1 (метод dot)
c = np.dot(a, b)
print(f"{c}\n")

# Находим их матричное произведение #2 (оператор @)
c = a @ b
print(f"{c}\n")
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
# Создаём случайный двумерный массив 7x7
a = np.random.randint(0, 49, size=(7, 7))
print(f"{a}\n")

# Находим у него среднее значение и стандартное отклонение
mean = np.mean(a)
std = np.std(a)
print(f"Среднее значение: {mean}\nСтандартное отклонение: {std}")
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
# ВАШ КОД ЗДЕСЬ
df.sort_values(by='total_bill', ascending=True)['total_bill'].plot(kind='hist', bins=20, ylabel = 'количество', xlabel='сумма', title='распределение total_bill')
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
# ВАШ КОД ЗДЕСЬ
sns.scatterplot(data=df, x="total_bill", y="tip")
plt.title("Взаимосвязь total_bill и tip")
plt.xlabel("total_bill")
plt.ylabel("tip")
plt.grid(alpha=0.3)
plt.show()
```

### Постройте pairplot

```python
# ВАШ КОД ЗДЕСЬ
sns.pairplot(df, hue='sex');
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
# ВАШ КОД ЗДЕСЬ
sns.scatterplot(data=df, x="total_bill", y="day")
plt.title("Взаимосвязь total_bill и day")
plt.xlabel("total_bill")
plt.ylabel("day")
plt.grid(alpha=0.3)
plt.show()
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
# ВАШ КОД ЗДЕСЬ
time_array = df['time'].unique()
time_count = len(time_array)

fig, axes = plt.subplots(1, time_count, figsize=(12, 5), sharey=True)

for ax, t in zip(axes, time_array):
    sns.histplot(data=df[df['time'] == t], x='tip', bins=12, kde=True, ax=ax)
    ax.set_title(f"Распределение {t}")
    ax.set_xlabel('tip')
    ax.set_ylabel('Количество')
    ax.set_xticks(np.arange(0, df['tip'].max() + 1, 1))

plt.tight_layout()
plt.show()
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
# ВАШ КОД ЗДЕСЬ
sex_array = df['sex'].unique()
sex_count = len(sex_array)

fig, axes = plt.subplots(1, sex_count, figsize=(12, 5), sharey=True)

for ax, s in zip(axes, sex_array):
    sns.scatterplot(data=df[df['sex'] == s], x="total_bill", y="tip", hue='smoker', ax=ax)
    ax.set_title(s)
    ax.set_xticks(np.arange(0, df['total_bill'].max() + 1, 5))
    ax.set_yticks(np.arange(0, df['tip'].max() + 1, 1))

    handles, labels = ax.get_legend_handles_labels()
    rename_map = {'0': 'No', '1': 'Yes'}

    new_labels = [rename_map.get(lbl, lbl) for lbl in labels]
    ax.legend(handles=handles, labels=new_labels, title='Smoker')

plt.tight_layout()
plt.show()
```
