## Задание 1: Зеркальное отражение цифр

Пользователь вводит пятизначное число. Программа должна зеркально отразить три центральные цифры. Первая и последняя остаются на местах.

<br>

**Примеры:**
* `23456 -> 25436`
* `30789 -> 38709`

```python
# ВАШ КОД ЗДЕСЬ
number = ""

while len(number) != 5 or not number.isnumeric():
    number = input("Введите пятизначное число!")

print(number + " -> " + number[0]+number[-2:-5:-1]+number[-1])
```

## Задание 2: Подсчет выходных до отпуска

Пользователь вводит сколько дней осталось до ближайшего отпуска. Программа должна вывести количество выходных дней до отпуска, если учесть, что выходные это суббота и воскресенье, сегодня понедельник и праздники мы не учитываем.  

<br>

**Примеры:**
* `4  -> 0`
* `6  -> 1`
* `14 -> 4`

```python
# ВАШ КОД ЗДЕСЬ
days_before = ""
while not days_before.isnumeric():
    days_before = input("сколько дней осталось до ближайшего отпуска: ")

print(int(days_before)//7*2 + int(days_before)%7//6)
```

## Задание 3: Плитка шоколада

Пользователь вводит длину и ширину плитки шоколада, а также размер куска, который хочет отломить. Программа должна вычислить - можно ли совершить подобный разлом или нет, если учесть, что ломать можно только по прямой.

<br>

**Примеры:**
* `3, 4, 6  -> True`
* `5, 7, 8  -> False`
* `4, 5, 12 -> True`

```python
# ВАШ КОД ЗДЕСЬ
ch_length = ""
ch_width = ""
ch_area = ""

while not ch_length.isnumeric():
    ch_length = input("Введите длину шоколадки в дольках: ")
while not ch_width.isnumeric():
    ch_width = input("Введите ширину шоколадки в дольках: ")
while not ch_area.isnumeric():
    ch_area = input("Введите количество долек, которое хотите отломить: ")

if int(ch_area) < int(ch_length)*int(ch_width) and (int(ch_area) % int(ch_length) == 0 or int(ch_area) % int(ch_width) == 0):
    print(ch_length + ", " + ch_width + ", " + ch_area + " -> True")
else:
    print(ch_length + ", " + ch_width + ", " + ch_area + " -> False")
```

## Задание 4: Римские числа

Пользователь вводит целое положительное число, программа должна вернуть строку в виде римского числа.

<br>

**Примеры:**
* `3   -> III`
* `15  -> XV`
* `234 -> CCXXXIV`

```python
# ВАШ КОД ЗДЕСЬ
number = ""
latin_num = ""
latin_num_array = [["I", "V"], ["X", "L"], ["C", "D"], ["M", "M"]]

while not number.isnumeric(): 
   number = input("Введите целое положительное число: ")[::-1]

for i in range(0, len(number)):
         if int(number[i]) <= 3:
            latin_num = latin_num_array[i][0] * int(number[i]) + latin_num
         elif int(number[i]) == 4:
            latin_num = latin_num_array[i][0] + latin_num_array[i][1] + latin_num
         elif int(number[i]) == 5:
            latin_num = latin_num_array[i][1] + latin_num
         elif int(number[i]) <= 8:
            latin_num = latin_num_array[i][1] + latin_num_array[i][0] * (int(number[i]) - 5) + latin_num
         elif int(number[i]) == 9:
            latin_num = latin_num_array[i][0] + latin_num_array[i+1][0] + latin_num
            
print(number[::-1] + " -> " + latin_num)
```

## Задание 5: Проверка на вещественное число

Пользователь вводит данные, проверить - являются ли они положительным вещественным числом. Не использовать встроенные функции для проверки, только методы данных и конструкцию IF. (Дополнительное задание, по желанию - проверка на отрицательные вещественные числа)

<br>

**Примеры:**
* `5.6  -> True`
* `.78  -> True`
* `.67. -> False`
* `5    -> True`

```python
# ВАШ КОД ЗДЕСЬ
data_num = input("Введите целое положительное число: ")

if data_num.isnumeric():
    print(data_num + " -> " + "True")
elif (data_num.replace('.', '', 1)).isnumeric() and not data_num.endswith('.'):
    print(data_num + " -> " + "True")
else:
    print(data_num + " -> " + "False")
```
