## Задание 1: Конвертер регистров

Написать функцию, которая будет переводить snake_case в PascalCase и наоборот. 

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

<br>

**Примеры:**
* `otus_course     -> OtusCourse`
* `PythonIsTheBest -> python_is_the_best`



```python
# ВАШ КОД ЗДЕСЬ
def notation_changer(str_txt: str) -> str:
    if "_" in str_txt:
        # Перевод из snake_case в PascalCase 
        lst_parts = str_txt.split("_")
        return ''.join(word.capitalize() for word in lst_parts)
    else:
        # Перевод из PascalCase в snake_case
        lst_result = []
        for char in str_txt:
            if char.isupper():
                lst_result.append('_' + char.lower())
            else:
                lst_result.append(char)
        return ''.join(lst_result).lstrip('_')

print("otus_course", notation_changer("otus_course"),sep=" => ")
print("PythonIsTheBest", notation_changer("PythonIsTheBest"),sep=" => ")
```

## Задание 2: Проверка валидности даты

Написать функцию проверяющую валидность введенной даты.

<br>

**Примеры:**
* `29.02.2000 -> True`
* `29.02.2001 -> False`
* `31.04.1962 -> False`



```python
# ВАШ КОД ЗДЕСЬ
str_date = input("Введите дату в формате dd.mm.yyyy: ")

def validate_date(date_str: str) -> bool:
    if len(date_str) != 10:
        return False
    parts = date_str.split(".")
    if len(parts) != 3:
        return False
    day, month, year = parts
    if not (day.isdigit() and month.isdigit() and year.isdigit()):
        return False
    day, month, year = int(day), int(month), int(year)
    if not (1 <= day <= 31 and 1 <= month <= 12 and year >= 0):
        return False
    if month == 2:
        if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
            if day > 29:
                return False
        else:
            if day > 28:
                return False
    elif month in [4, 6, 9, 11] and day > 30:
        return False
    return True
        
print(str_date, validate_date(str_date), sep=" => ")
```

## Задание 3: Проверка на простое число

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

<br>

**Примеры:**
* `17 -> True`
* `20 -> False`
* `23 -> True`

```python
# ВАШ КОД ЗДЕСЬ
str_num = input("Введите простое число: ")

def is_prime(num_str: str) -> bool:
    if not num_str.isdigit():
        return False
    num = int(num_str)
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

print(str_num, is_prime(str_num), sep=" => ")
```

## Задание 4: Учет пользователей

Пользователь в бесконечном цикле вводит данные пользователей: имя, затем фамилию, возраст и ID. Ввод продолжается до тех пор, пока не будет введено пустое поле. 

Пользователи заносятся в словарь, где ключ это ID пользователя, а остальные данные записываются в виде кортежа. 

**Программа должна проверять:**
* имя и фамилия состоят только из символов и начинаются с большой буквы - если не с большой, то заменяет букву на большую;
* возраст должен быть числом от 18 до 60;
* ID - целое число, дополненное до 8 знаков незначащими нулями, ID должен быть уникальным.

**Дополнительно:** написать функцию, которая будет выводить полученный словарь в виде таблицы.

```python
# ВАШ КОД ЗДЕСЬ

dct_record = {}

def set_upper_alpha(str_txt: str) -> str:
    if not str_txt.isalpha():
        return ""
    else:
        return str_txt.capitalize()
    
def check_age(str_age: str) -> int:
    if not str_age.isdigit():
        return 0
    age = int(str_age)
    if not 18 <= age <= 60:
        return 0
    return age

def check_id(str_id: str) -> str:
    if not str_id.isdigit() and len(str_id) > 8:
        return ""
    else:
        return "0" * (8 -len(str_id)) + str_id

def print_record(dct_rec: dict) -> None:
    imia_max = max(len(value[0]) for value in dct_rec.values())
    familija_max = max(len(value[1]) for value in dct_rec.values())

    for key, value in dct_rec.items():
        imia, familija, vozrast = value
        print(f"{key} {imia} {" " * (imia_max - len(imia))} {familija} {" " * (familija_max - len(familija))} {vozrast}")
    
while True:
    str_imia = set_upper_alpha(input("Имя: "))
    if str_imia == "":
        break
    str_familija = set_upper_alpha(input("Фамилия: "))
    if str_familija == "":
        break
    str_vozrast = check_age(input("Возраст: "))
    if str_vozrast == 0:
        break
    str_id = check_id(input("ID: "))
    if str_id == "":
        break
    elif str_id in dct_record:
        print(f"Запись с ID {str_id} уже существует.")
        continue
    else:
        dct_record[str_id] = (str_imia, str_familija, str_vozrast)
        print_record(dct_record)
```
